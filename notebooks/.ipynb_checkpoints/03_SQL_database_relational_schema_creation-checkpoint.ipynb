### CREATING NORMALIZED SQL RELATIONAL SCHEMA

#### Purpose:
To demonstrate how to design and build a fully normalized SQL relational schema for chromatographic quality control analytics involving HPLC and GC instrumentationS.

#### Objectives:
- Design a clean 3NF relational schema suitable for chromatographic QC dataset.   
- Implement foreign keys and indexing.

#### Why This Step Matters:
In production laboratory environments, QC data is typically stored in relational databases rather than spreadsheets. Therefore i structured the QC data using SQL, loaded the cleaned dataset into a SQL database which emulates real-world data infrastructure and enables structured querying, aggregation, and traceability. SQL-based summaries support routine QC reporting, instrument comparison, and historical performance review.


The created schema contains six core tables with various columns as indicated below:

#### 1. instrument:
metadata for HPLC_01, GC_02 (instrument_id, model, manufacturer, location)
                                 
#### 2. samples:
sample_id, instrument_id, retention_time_min, peak_area, peak_width_min, concentration_mgL, 
true_value_mgL, run_date, month
    
#### 3. sample_metrics:
sample_id, instrument_id, error_mgL, error_pct, residual, percent_recovery, response_factor,
peakarea_zscore, retention_deviation, is_outlier, rsd_pct, replicate_number
    
#### 4. calibrations:
instrument_id, slope, intercept, r2
    
#### 5. control_summary:
instrument_id, total_runs, out_of_control, out_of_control_runs, mean_peak_area, std_peak_area, ucl, lcl
    
#### 6. system_suitability:
instrument_id, sample_id,	plates,	resolution,	tailing 

This design eliminates redundancy, supports many-to-one relationships, and allows 
robust QC analytics.
