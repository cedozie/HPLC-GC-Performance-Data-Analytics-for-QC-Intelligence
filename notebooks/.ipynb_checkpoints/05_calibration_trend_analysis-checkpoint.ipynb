#### Abstract
This report evaluates chromatographic calibration performance using parity
analysis, accuracy trending, response factor stability, and linearity assessment.
Statistical quality control tools were applied to demonstrate continued
instrument suitability, detect emerging bias or drift, and support objective
fitness-for-use decisions in alignment with ICH Q2(R2) and USP <1225>.


### Experimental Context and QC Objectives

Calibration verification is a critical component of analytical method
performance monitoring. Under ICH Q2(R2), calibration data must demonstrate
acceptable linearity, accuracy, and stability throughout routine operation.

The objectives of this analysis were to:
- Confirm agreement between measured and true concentrations
- Monitor accuracy and recovery trends over time
- Evaluate response factor stability using statistical control limits
- Verify calibration linearity against predefined acceptance criteria
- Integrate all metrics into an objective instrument fitness assessment


```python
# ============================ SETUP AND DATA PATH =================================
import os
import sys
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sqlalchemy import create_engine
from datetime import datetime

sns.set(style="whitegrid", palette="muted", font_scale=1.1)

# Output paths
OUTPUT_DIR = "calibration_trend_analysis_outputs"
FIG_DIR = f"{OUTPUT_DIR}/figures"
TABLE_DIR = f"{OUTPUT_DIR}/tables"
for d in [OUTPUT_DIR, FIG_DIR, TABLE_DIR]:
    os.makedirs(d, exist_ok=True)

# Database connection
db_path = "/Users/christopheredoziesunday/Documents/Chrom-Data-Analytics/qc_structured.db"
engine = create_engine(f"sqlite:///{db_path}")

# Load dataset
df = pd.read_sql("""
    SELECT s.sample_id, s.instrument_id, s.concentration_mgL,
           s.true_value_mgL, s.run_date, s.month,
           
           sm.percent_recovery, sm.response_factor,
           sm.is_outlier, c.slope, c.intercept, c.r2
    
    FROM samples s
    LEFT JOIN sample_metrics sm ON s.sample_id = sm.sample_id
    LEFT JOIN calibrations c ON s.instrument_id = c.instrument_id
""", con=engine)

print(f"Dataset loaded: {df.shape[0]} records")
```

### Data Integrity and Pre-Processing

Calibration records were validated for completeness, temporal consistency, and
instrument attribution prior to statistical evaluation. Ensuring reliable
time-ordering and removal of incomplete records is essential for trend and
stability analyses.


```python
df["run_date"] = pd.to_datetime(df["run_date"], errors="coerce")
df = df.dropna(subset=["run_date", "instrument_id"])
df = df.sort_values("run_date")
instrs = df["instrument_id"].unique()

print(f"Instruments evaluated: {len(instrs)}")
```

### QC Acceptance Criteria

- Percent Recovery: 95–105%
- Calibration Linearity: R² ≥ 0.995
- Response Factor Stability: CV ≤ 5%


```python
QC_LIMITS = {
    "recovery_low": 95,
    "recovery_high": 105,
    "r2_min": 0.995,
    "response_factor_cv_max": 5.0  # %
}
```

### Calibration Parity Assessment

Calibration accuracy and proportionality were evaluated using parity plots
comparing measured concentrations to certified true values across instruments.
Data points clustering along the 1:1 parity line indicate acceptable agreement
between measured and nominal concentrations. Systematic deviation from the
parity line may suggest proportional bias, calibration slope error, or matrix
effects requiring further investigation.


```python
# ============================ FIGURE 1: PARITY PLOT ============================
g = sns.lmplot(
    data=df,
    x="true_value_mgL",
    y="concentration_mgL",
    hue="instrument_id",
    height=6,
    aspect=1.2,
    scatter_kws={"alpha": 0.7}
)

# 1:1 parity line
ax = g.ax
lims = [min(ax.get_xlim()[0], ax.get_ylim()[0]), max(ax.get_xlim()[1], ax.get_ylim()[1])]
ax.plot(lims, lims, "--k", linewidth=1)
ax.set_xlim(lims)
ax.set_ylim(lims)

plt.title("Figure 1. Calibration Parity Plot by Instrument", weight="bold")
plt.tight_layout()
plt.savefig(f"{FIG_DIR}/figure_1_parity_plot.png", dpi=300)
plt.show()
```

### Calibration Accuracy and Recovery Trends

Calibration accuracy was assessed using mean percent recovery grouped by month
and instrument. Acceptance was defined as percent recovery within the range of
95–105%. The heatmap visualization enables rapid identification of temporal
trends, instrument-specific bias, and potential drift in analytical performance
over time.


```python
# =============== FIGURE 2: ACCURACY HEATMAP (% RECOVERY OVER TIME) ================
heatmap_data = (df.groupby(["month", "instrument_id"])["percent_recovery"].mean().unstack())

plt.figure(figsize=(10,6))
sns.heatmap(
    heatmap_data,
    annot=True,
    fmt=".1f",
    cmap="coolwarm",
    center=100,
    vmin=QC_LIMITS["recovery_low"],
    vmax=QC_LIMITS["recovery_high"],
    linewidths=0.5
)

plt.title("Figure 2. Mean Percent Recovery by Month and Instrument", weight="bold")
plt.tight_layout()
plt.savefig(f"{FIG_DIR}/figure_2_accuracy_heatmap.png", dpi=300)
plt.show()
```

### Response Factor Stability Evaluation

Westgard-style control response factor stability was evaluated over time for each instrument to assess calibration robustness and detector consistency. Stable response factors indicate reliable analytical performance, whereas increasing variability or systematic drift may reflect detector aging, source contamination, changes in chromatographic conditions or maintenance requirements. Stability was assessed relative to predefined QC acceptance criteria.


```python
# ========== FIGURE 3: WESTGARD CONTROL CHART – RESPONSE FACTOR ==========
def westgard_limits(series):
    mean = series.mean()
    sd = series.std()
    return mean, mean - 2*sd, mean + 2*sd

plt.figure(figsize=(11,6))

for inst, d in df.groupby("instrument_id"):
    mean, l2s, u2s = westgard_limits(d["response_factor"])
    plt.plot(d["run_date"], d["response_factor"], marker="o", label=inst)
    plt.axhline(mean, linestyle="-", linewidth=1)
    plt.axhline(l2s, linestyle="--", linewidth=0.8)
    plt.axhline(u2s, linestyle="--", linewidth=0.8)

plt.title("Figure 3. Response Factor Stability (Westgard Monitoring)", weight="bold")
plt.xlabel("Run Date")
plt.ylabel("Response Factor")
plt.legend(title="Instrument")
plt.tight_layout()
plt.savefig(f"{FIG_DIR}/figure_3_response_factor_stability.png", dpi=300)
plt.show()
```

### Calibration Linearity Assessment (R²)
Calibration linearity was assessed using the coefficient of determination (R²).
Acceptance was defined as R² ≥ 0.995, consistent with chromatographic method
validation expectations. This assessment confirms proportionality across the
analytical range.


```python
# ========================== FIGURE 4: CALIBRATION LINEARITY (R²) ==========================
# Extract one R² value per instrument
r2_by_instrument = (df.groupby("instrument_id")["r2"].first().reset_index())

plt.figure(figsize=(10,6))
plt.bar(r2_by_instrument["instrument_id"],r2_by_instrument["r2"])
# QC acceptance line
plt.axhline(
    QC_LIMITS["r2_min"],
    linestyle="--",
    linewidth=1,
    label=f"Acceptance Limit (R² ≥ {QC_LIMITS['r2_min']})"
)

plt.title("Figure 4. Calibration Linearity (R²) per Instrument", weight="bold")
plt.xlabel("Instrument ID")
plt.ylabel("R²")
plt.ylim(0.95, 1.00)
plt.legend()
plt.tight_layout()
plt.savefig(f"{FIG_DIR}/figure_4_calibration_linearity_r2.png", dpi=300)
plt.show()
```

```python
# ========================== TABLE 1: QC SUMMARY + INSTRUMENT FITNESS =========================
qc_fitness_summary = (
    df.groupby("instrument_id")
      .agg(
          mean_recovery=("percent_recovery", "mean"),
          recovery_sd=("percent_recovery", "std"),
          response_factor_cv=("response_factor", lambda x: (x.std() / x.mean()) * 100),
          min_r2=("r2", "min"),
          n_samples=("sample_id", "count")
      )
      .reset_index()
)

# Apply QC acceptance criteria
qc_fitness_summary["recovery_pass"] = qc_fitness_summary["mean_recovery"].between(
    QC_LIMITS["recovery_low"], QC_LIMITS["recovery_high"]
)

qc_fitness_summary["r2_pass"] = qc_fitness_summary["min_r2"] >= QC_LIMITS["r2_min"]
qc_fitness_summary["rf_cv_pass"] = qc_fitness_summary["response_factor_cv"] <= QC_LIMITS["response_factor_cv_max"]

qc_fitness_summary["overall_qc_pass"] = (
    qc_fitness_summary["recovery_pass"] &
    qc_fitness_summary["r2_pass"] &
    qc_fitness_summary["rf_cv_pass"]
)


qc_fitness_summary["recovery_score"] = np.clip(
    1 - abs(qc_fitness_summary["mean_recovery"] - 100) / 10, 0, 1
)

qc_fitness_summary["r2_score"] = np.clip(
    (qc_fitness_summary["min_r2"] - 0.99) / (1 - 0.99), 0, 1
)

qc_fitness_summary["rf_cv_score"] = np.clip(
    1 - qc_fitness_summary["response_factor_cv"] / QC_LIMITS["response_factor_cv_max"], 0, 1
)

qc_fitness_summary["fitness_index"] = (
    0.30 * qc_fitness_summary["recovery_score"] +
    0.30 * qc_fitness_summary["r2_score"] +
    0.25 * qc_fitness_summary["rf_cv_score"]
)

qc_fitness_summary["fitness_class"] = pd.cut(
    qc_fitness_summary["fitness_index"],
    bins=[0, 0.7, 0.85, 1.0],
    labels=["Not Fit", "Marginal", "Fit"]
)

qc_fitness_summary

qc_fitness_summary.to_csv(f"{TABLE_DIR}/table_1_qc_instrument_fitness_summary.csv", index=False)
```

### Results and Discussion

Calibration performance across the instruments demonstrated strong agreement between
measured and true concentrations, as evidenced by parity analysis (Figure 1).
Accuracy remained within predefined acceptance limits for most instruments,
although temporal trends highlighted potential bias in select cases (Figure 2).

Response factor stability monitoring confirmed consistent detector performance
with limited variability over time (Figure 3). Linearity assessment further
supported calibration robustness, with R² values exceeding ICH Q2(R2)
expectations across instruments (Figure 4).

Integrated evaluation of all calibration metrics supports continued suitability
of the analytical systems for quantitative use.


### QC Conclusions and Instrument Fitness

Instruments meeting recovery, linearity, and response factor stability criteria
were deemed fit for continued analytical use. Instruments exhibiting marginal or
failing performance should be reviewed for recalibration, maintenance, or
requalification in accordance with USP <1225> guidance.
